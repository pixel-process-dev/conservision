## Create Full Image Meta Data

Creates a data directory of the full images that matches setup of cropped images use for most models.

```python
import pandas as pd
import os
```

```python
train_data = pd.read_csv('../data/competition/train_features.csv')
train_data['split'] = "train"
```

```python
train_labels = pd.read_csv('../data/competition/train_labels.csv')
```

```python
test_data = pd.read_csv('../data/competition/test_features.csv')
test_data['split'] = "test"
```

```python
combined = pd.concat([train_data, test_data])
```

```python
combined.shape
```

```python
with_labels = pd.merge(combined, train_labels, on='id', how="outer")
```

```python
with_labels.shape
```

```python
with_labels.isnull().sum()
```

```python
with_labels['original_id'] = with_labels['id'].copy()
```

```python
with_labels['crop_filename'] = with_labels['filepath'].apply(os.path.basename)
```

```python
with_labels.columns
```

```python
with_labels.to_csv('../data/competition/full_images/crop_metadata.csv', index=False)
```
